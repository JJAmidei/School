```python
# Import required modules 
import numpy as np 
import matplotlib.pyplot as plt 
  
# 1D arrays 
x = np.arange(-5,5,0.1) 
y = np.arange(-5,5,0.1) 
  
# Meshgrid 
X,Y = np.meshgrid(x,y) 
  
# Assign vector directions 
Ex = (X + 1)/((X+1)**2 + Y**2) - (X - 1)/((X-1)**2 + Y**2) 
Ey = Y/((X+1)**2 + Y**2) - Y/((X-1)**2 + Y**2) 
  
# Depict illustration 
plt.figure(figsize=(10, 10)) 
plt.streamplot(X,Y,Ex,Ey, density=1.4, linewidth=None, color='#A23BEC') 
plt.plot(-1,0,'-or') 
plt.plot(1,0,'-og') 
plt.title('Electromagnetic Field') 
  
# Show plot with grid 
plt.grid() 
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
  
# Define the range for plotting
x = np.arange(-10, 10, 0.1) 
y = np.arange(-10, 10, 0.1) 
  
# Meshgrid 
X, Y = np.meshgrid(x, y) 

# Define the initial position and velocity of the charge
initial_position = np.array([0, 0])
initial_velocity = np.array([1, 0])  # Unit velocity along x-axis

# Define the acceleration function
def acceleration(t):
    return np.array([0.1, 0])  # Constant acceleration in x-direction

# Define the time step and total time
dt = 0.1
total_time = 10

# Simulate the motion and calculate the electric field at each time step
positions = [initial_position]
velocities = [initial_velocity]
times = np.arange(0, total_time, dt)
for t in times[1:]:
    a = acceleration(t)
    new_velocity = velocities[-1] + a * dt
    new_position = positions[-1] + new_velocity * dt
    positions.append(new_position)
    velocities.append(new_velocity)

# Calculate the electric field at each position
def calculate_electric_field(position):
    r_squared = (X - position[0])**2 + (Y - position[1])**2
    Ex = (X - position[0]) / r_squared 
    Ey = (Y - position[1]) / r_squared
    return Ex, Ey

# Calculate the electric field at each position over time
electric_fields = [calculate_electric_field(pos) for pos in positions]

# Plot the electric field at a specific time
plt.figure(figsize=(10, 10)) 
Ex, Ey = electric_fields[-1]  # Electric field at the final time
plt.streamplot(X, Y, Ex, Ey, density=1.0, linewidth=None) 

# Plot the trajectory of the charge
trajectory_x = [pos[0] for pos in positions]
trajectory_y = [pos[1] for pos in positions]
plt.plot(trajectory_x, trajectory_y, 'r-')

# Title and grid
plt.title('Electric Field of an Accelerated Charge') 
plt.grid() 
plt.show()
```
