````python
def factorial(n): # Function that allows us to find the factorial of a number. 
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)
    
factorial(500)
````

As a final note, here's a plot. 

````python
import matplotlib.pyplot as plt
import numpy as np

physics = [0.5, 1, 1.5, 2, 2.5]
coding = [10, 20, 50, 100, 200]

fig, ax = plt.subplots()

plt.scatter(physics, coding, label='Data Points', color='blue')

params = np.polyfit(physics, np.log(coding), 1)
a, b = params
y_fit = np.exp(a * np.array(physics) + b)

plt.plot(physics, y_fit, label=f'Exponential Fit (y = e^{a:.2f}x)', color='red')

plt.xlabel('Level of physics (year)')
plt.ylabel('Amount of coding (minutes per day)')
plt.title('Level of physics vs. Amount of coding I have to do')
plt.legend()

plt.grid(True)
plt.show()
````
