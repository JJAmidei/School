```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,2*np.pi,0.1)
y1 = np.cos(2*x)
y2 = np.sin(x)

plt.figure()
plt.plot(x, y1, color='r')
plt.plot(x, y2, color='b')
plt.plot()
plt.grid(True)
plt.show()
```
