Using a systm of units in which the electron mass $m=1$ and $\hbar=1$, an electron in a potential $V(z) = z^2/2$ has a wavefunction at a given instant in time 

$$\psi(z) = \frac{1}{\sqrt{2\sqrt{\pi}}} \biggl(1+\sqrt{2}z\biggr)e^{-z^2/2}$$

What is the expectation value of the energy for the particle in this state?

#### Time-independent Schrodinger equation in one-dimension

$$\biggr(-\frac{\hbar^2}{2 m_0}\frac{\partial^2}{\partial z^2} + V \biggl)\psi = E\psi $$

#### Hamiltonian 

$$ \hat{H} = -\frac{\hbar^2}{2 m_0}\frac{\partial^2}{\partial z^2} + V $$

$$\rightarrow \hat{H}\psi = E \psi$$

#### Expectation value of the energy 

$$ \langle E \rangle = \int^{\infty}_{-\infty} \psi^* \hat{H} \psi dz$$

#### Approximating the expectation value of the energy 

$$ \langle E \rangle = \int^{\infty}_{-\infty} \psi^* \hat{H} \psi dz \rightarrow \lim_{\Delta z\rightarrow 0} \sum^{\infty}_{n=-\infty}\psi^*_n \hat{H} \psi_n \Delta z$$

$$\rightarrow \psi^*_n \hat{H} \psi_n = \psi^*_n \biggr( -\frac{\hbar^2}{2 m_0}\frac{\partial^2}{\partial z^2} + V \biggl)\psi_n = -\frac{\hbar^2}{2 m_0}\psi^*_n\frac{\partial^2\psi_n}{\partial z^2} + V\psi^*_n\psi_n$$

Let $\psi_{n+1} - \psi_n = \psi(z + \Delta z) - \psi(z)$

$$ \frac{\partial\psi_n}{\partial z} = \lim_{\Delta z \rightarrow 0} \frac{\psi_{n+1} - \psi_n}{\Delta z} \rightarrow  \frac{\Delta \psi_n}{\Delta z} = \frac{\psi_{n+1} - \psi_n}{\Delta z} $$

$$ \frac{\partial^2 \psi_n}{\partial z^2} = \frac{\partial}{\partial z^2}\frac{\partial \psi_n}{\partial z} \rightarrow \frac{\Delta}{\Delta z}\frac{\Delta \psi_n}{\Delta z} = \frac{(\psi_{n+2} - \psi_{n+1}) - (\psi_{n+1} - \psi_{n})}{\Delta z^2} = \frac{(\psi_{n+2} - 2\psi_{n+1} + \psi_{n})}{\Delta z^2}$$

 $$\langle E \rangle = \sum^{\infty}_{n=-\infty}-\frac{\hbar^2}{2 m_0}\psi^*_n \biggl( \frac{\psi_{n+2} - 2\psi_{n+1} + \psi_{n}}{\Delta z^2} \biggr) + V\psi^*_n\psi_n \Delta z$$
 
#### Defining additional terms

$$\psi(z = 0) = \psi_0 =  \frac{1}{\sqrt{2\sqrt{\pi}}} \biggl(1+\sqrt{2}(0)\biggr)e^{-(0)^2/2} = \frac{1}{\sqrt{2\sqrt{\pi}}}(1)(1) = \frac{1}{\sqrt{2\sqrt{\pi}}}$$


$$\psi(z) = \psi_0\biggl(1+\sqrt{2}z\biggr)e^{-z^2/2}$$

$$V(z) = \frac{z^2}{2}$$

$$\psi^*_n\hat{H}\psi_n = -\frac{\hbar}{2 m_0}\psi^*_n\biggl( \frac{\psi_{n+2} - 2\psi_{n+1} + \psi_{n}}{\Delta z^2}\biggr) + \frac{z^2}{2}\psi^*_n\psi_n$$

Since we are told to use natural units, $\hbar = m = 1$. Additionally, since the wavefunction $\psi(z)$ is purely real, $\psi^* = \psi$, we can write the following. 

$$\psi^*_n\hat{H}\psi_n = -\frac{1}{2}\psi_n\biggl( \frac{\psi_{n+2} - 2\psi_{n+1} + \psi_{n}}{\Delta z^2}\biggr) + \frac{z^2}{2}\psi^2_n$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#%time # this is the only line I added inbetween the canvas submission and now 
hbar = 1 # In natural units
m = 1 # We were told to use these values in the book

domain = 100 # Half the total domain. This leads to us looking at z =[-domain, domain]
delta_z = 0.001 # Step size
z = np.arange(-domain, domain+delta_z, delta_z) # The domain+delta_z term is to keep both sides of the 
                                                # domain symmetric about 0. 

psi_0 = 1/(np.sqrt(2*np.sqrt(np.pi)))
psi = psi_0*(1+np.sqrt(2)*z)*np.e**(-z**2/2)

dpsi = np.zeros(len(psi)) # Creating empty array to store values of the wavefunction's 1st derivative 
dpsi[-1] = 0 # Defining final value as 0 because the loop below will create an array with 1 less entry than psi
             # We will need psi and dpsi to have the same size so we can eventually make d2psi (the 2nd derivative of psi
             # which is defined below) and psi the same size in order to add their multiples later. 

for n in np.arange(len(psi)-1):
    dpsi[n] = (psi[n+1] - psi[n])/delta_z
    
d2psi = np.zeros(len(psi))
d2psi[-1] = 0 # Same as above. We will need d2psi to have the same size as psi in order to add their multiples later.

for n in np.arange(len(psi)-1):
    #d2psi[n] = (psi[n+2] - 2*psi[n+1] + psi[n])/delta_z**2
    # I decided against using the top because it returned substantially more
    # error than the following line.
    d2psi[n] = (dpsi[n+1] - dpsi[n])/delta_z
    
V = z**2/2

I = -hbar/(2*m)*psi*d2psi + V*psi**2 # The integrand we use to find the energy expectation value.
                                     # This is defined above as the complex conjugate of the wavefunction
                                     # time the product of the Hamiltonian and the wavefunction. 
                                     # Since our wavefunction is real, this is effectively the same thing
                                     # as multiplying the Hamilontian by the wavefunction twice.
                
eEnergy = np.sum(I)*delta_z # expectation energy

print(f'<E> = {eEnergy} in natural units')
```

```python
# ------- Graphing Wave Function and Probability Density -------

psi_sq = np.abs(psi)**2

zero = len(psi)/2 - 0.5 # zero is centered at one-half less than the true halfway point
num2bin = len(z)/len(np.arange(-domain, domain+1, 1)) # converts from numbers into bins
interval = 5 # How far +/- from the origin you'd like to see 

lowerbound = int(np.round(zero - interval*num2bin))
upperbound = int(np.round(zero + interval*num2bin))

print(f'Number of bins: {len(psi)}')
print(f'z over the interval {-100, 100}')
print(f'Ratio of bins/z = {num2bin} ')
print(f'Zero is located at bin number {int(zero)}')

zero = len(psi)/2 - 0.5 # zero is centered at the halfway point
num2bin = len(z)/len(np.arange(-domain, domain+1, 1)) # converts from numbers into bins
interval = 10 # How far +/- from the original

fig, (ax1, ax2) = plt.subplots(2,1, figsize=(10,6))

ax1.plot(z[lowerbound:upperbound], psi[lowerbound:upperbound], 'r')
ax1.set_xlabel('z')
ax1.set_ylabel(r'$\psi$')
ax1.grid()
ax1.set_title(r'Wavefunction $\psi(z)$')

ax2.plot(z[lowerbound:upperbound], psi_sq[lowerbound:upperbound], 'b')
ax2.set_xlabel('z')
ax2.set_ylabel(r'$|\psi|^2$')
ax2.grid()
ax2.set_title(r'Probability Density $|\psi(z)|^2$')

plt.tight_layout()
```
