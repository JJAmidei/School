```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
hbar = 1  # Set to 1 for simplicity
omega_tot = 1  # Set to a suitable value for omega_tot

# Time array
t = np.linspace(0, 10, 1000)  # Change 10 to a larger value for longer time

# Magnetic Field 1 (B_0 = 2T, B_1 = 1T)
theta_1 = np.radians(26.6)
Sx1 = hbar * np.sin(theta_1) * np.sin(omega_tot * t / 2)
Sz1 = hbar * (np.cos(theta_1)**2 - 0.2 * np.sin(omega_tot * t / 2)**2)

# Magnetic Field 2 (B_0 = 1T, B_1 = 1T)
theta_2 = np.radians(45)
Sx2 = hbar * np.sin(theta_2) * np.sin(omega_tot * t / 2)
Sz2 = hbar * (np.cos(theta_2)**2 - 0.5 * np.sin(omega_tot * t / 2)**2)

# Setting Sy to zero
Sy1 = np.zeros_like(t)
Sy2 = np.zeros_like(t)

# Plotting
plt.figure(figsize=(12, 8))

# Sx plots
plt.subplot(3, 1, 1)
plt.plot(t, Sx1, label='Sx (B0=2T, B1=1T)', color='blue')
plt.plot(t, Sx2, label='Sx (B0=1T, B1=1T)', color='orange')
plt.title('Expectation Values of Spin Components')
plt.ylabel('$\\langle \\hat{S}_x \\rangle$')
plt.legend()
plt.grid()

# Sy plots
plt.subplot(3, 1, 2)
plt.plot(t, Sy1, label='Sy (B0=2T, B1=1T)', color='blue', linestyle='--')
plt.plot(t, Sy2, label='Sy (B0=1T, B1=1T)', color='orange', linestyle='--')
plt.ylabel('$\\langle \\hat{S}_y \\rangle$')
plt.legend()
plt.grid()

# Sz plots
plt.subplot(3, 1, 3)
plt.plot(t, Sz1, label='Sz (B0=2T, B1=1T)', color='blue')
plt.plot(t, Sz2, label='Sz (B0=1T, B1=1T)', color='orange')
plt.ylabel('$\\langle \\hat{S}_z \\rangle$')
plt.xlabel('Time')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
hbar = 1  # Set to 1 for simplicity
omega_tot = 1  # Set to a suitable value for omega_tot

# Time array
t = np.linspace(0, 10, 1000)  # Change 10 to a larger value for longer time

# Magnetic Field 1 (B_0 = 2T, B_1 = 1T)
theta_1 = np.radians(26.6)
Sx1 = hbar * np.sin(theta_1) * np.sin(omega_tot * t / 2)
Sz1 = hbar * (np.cos(theta_1)**2 - 0.2 * np.sin(omega_tot * t / 2)**2)

# Magnetic Field 2 (B_0 = 1T, B_1 = 1T)
theta_2 = np.radians(45)
Sx2 = hbar * np.sin(theta_2) * np.sin(omega_tot * t / 2)
Sz2 = hbar * (np.cos(theta_2)**2 - 0.5 * np.sin(omega_tot * t / 2)**2)

# Sy is zero
Sy1 = np.zeros_like(t)
Sy2 = np.zeros_like(t)

# Plotting the expectation vector in 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# For Magnetic Field 1
ax.quiver(np.zeros_like(Sx1), np.zeros_like(Sy1), np.zeros_like(Sz1), 
          Sx1, Sy1, Sz1, color='blue', label='B0=2T, B1=1T', alpha=0.5)

# For Magnetic Field 2
ax.quiver(np.zeros_like(Sx2), np.zeros_like(Sy2), np.zeros_like(Sz2), 
          Sx2, Sy2, Sz2, color='orange', label='B0=1T, B1=1T', alpha=0.5)

# Set the limits
ax.set_xlim([-hbar, hbar])
ax.set_ylim([-hbar, hbar])
ax.set_zlim([-hbar, hbar])

# Set labels and title
ax.set_xlabel('$\\langle \\hat{S}_x \\rangle$')
ax.set_ylabel('$\\langle \\hat{S}_y \\rangle$')
ax.set_zlabel('$\\langle \\hat{S}_z \\rangle$')
ax.set_title('Expectation Value Vectors of Spin in 3D Space')
ax.legend()

plt.show()
```
